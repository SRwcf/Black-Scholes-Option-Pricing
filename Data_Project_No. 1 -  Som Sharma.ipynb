#Question_No. 2:Implementing n-period recombining BTM & programming it to price a European style derivative with a Payoff Function.



```python
def Option_Pricing(S0, T, n, r, K, volatility, option):
    dt = T/n
    u = np.exp(volatility*np.sqrt(dt))
    d = 1/u
    p = .5*(1+(((r/volatility)-(volatility/2))*np.sqrt(dt)))
    discount = 1/(1 + r*dt)
    
    
    # Stock values at maturity
    
    S = np.zeros(n+1)
    S[0] = S0*d**n
    for i in range(1, n+1):
        S[i] = S[i-1]*u/d
     
    
    # Option payoffs at maturity
    
    V = np.zeros(n+1)
    
    if option == "Call":
        for i in range(0, n+1):
            V[i] = max(0, S[i] - K)
            
    if option == "Put":
        for i in range(0, n+1):
            V[i] = max(0, K - S[i])
            
    if option == "Forward":
        for i in range(0, n+1):
            V[i] = S[i] - K
            
            
    # Option's Price today
    
    for j in np.arange(n, 0, -1):
        for i in range(0, j):
            V[i] = discount*(p*V[i+1] + (1-p)*V[i])
            
            
    return V[0]
```

```python
# Checking_Models_Generability

# Call option

Call_Option_Price = Option_Pricing(S0=100, T=0.0833, n=4, r=0.0006, K=100, volatility=0.1, option="Call")
print(Call_Option_Price)

# Put option

Put_Option_Price = Option_Pricing(S0=100, T=0.0833, n=4, r=0.0006, K=100, volatility=0.1, option="Put")
print(Put_Option_Price)

# Forward contract

Forward_Option_Price = Option_Pricing(S0=100, T=0.0833, n=4, r=0.0006, K=100, volatility=0.1, option="Forward")
print(Forward_Option_Price)
```

#OBSERVATION AND SUMMARY:

When the volatility of the stock rises,the prices for the call and put options substantially increase.

While checking the output from the model above, the model seems to implement an n-period recombining BTM as well as
price a European style derivative with a same payoff function F(s) with reasonable level of accuracy.



#Question_No. 3 : n-period recombining BTM Option Pricing vs Black-Scholes Option Pricing



```python
def d1(S0,T,r,K,volatility):
    return(log(S0/K)+(r+volatility**2/2.)*T)/(volatility*sqrt(T))
def d2(S0,T,r,K,volatility):
    return d1(S0,T,r,K,volatility)-volatility*sqrt(T)

def bs_call(S0,T,r,K,volatility):
    return S0*norm.cdf(d1(S0,T,r,K,volatility))-K*exp(-r*T)*norm.cdf(d2(S0,T,r,K,volatility))
  
def bs_put(S0,T,r,K,volatility):
    return K*exp(-r*T)-S0+bs_call(S0,T,r,K,volatility)
```
